## 1) Importamos las librerías necesarias 

```python
#Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importamos los datos
df=pd.read_csv('Accidentes_tratado.csv')
```

### 2.1) Accidentes por país

```python
#Considero agrupar por año los países con accidentes y contarlos
accidents_per_country = df.groupby('country')['year'].count()
```

```python
#Elijo no considerar los países = other ya que no tengo información certera de la locación
accidents_per_country  = accidents_per_country.drop('other', errors='ignore')
```

```python
#Me quedo con los 10 valores más altos
top_10_accidents_per_country=accidents_per_country.sort_values(ascending=False).head(10)
```

```python
#Verifico los datos obtenidos
top_10_accidents_per_country
```

```python
#Graficamos
plt.figure(figsize=(10, 6))  
sns.barplot(x=top_10_accidents_per_country.values, y=top_10_accidents_per_country.index, palette="Blues_r", linewidth=0.3)
plt.xlabel('Number of Accidents') 
plt.ylabel('Country')
plt.title('Number of Accidents by Country')  
plt.show()
```

Podemos concluir que 1042 de los 5008 accidentes fueron en Estados Unidos, es decir, que el 20.8% de los accidentes se encuentran nucleados en Estados Unidos. <br>

###  2.7) Accidentes por categoría de aeronaves: militares o de pasajeros

### 3.3) Matriz de correlación

Vamos a crear una matriz de correlación para ver el grado de correlación entre mis variables numéricas.

```python
#Creamos la matriz de correlación
df_num = df[['all_aboard','passengers_aboard','crew_aboard','total_fatalities','passenger_fatalities','crew_fatalities', 'survivors']]
corr_matrix = df_num.corr()
plt.figure(figsize=(10, 8))  
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
```

Podemos observar que hay una amplia correlación entre las variables, al aumentar el número de total de personas abordo, aumento el número de pasajeros y de tripulación abordo, lo mismo si aumento el total_fatalities, se ve afectado el passenger fatality y crew fatalities. 
La menor corelación tiene que ver entre el número de muertos y el número de sobrevivientes, lo que indica que no hay relación entre las variables.

#### 3.9.1) Comparación fallecidos por año y personas a bordo por año

```python
# Agrupamos por año y sumar las columnas necesarias
comparison_total_fatalities = df.groupby(df['year'])[['total_fatalities', 'all_aboard']].sum()
```

```python
comparison_total_fatalities
```

```python
plt.figure(figsize=(18, 6))
plt.subplot(1, 2, 1)
plt.fill_between(comparison_total_fatalities.index, 'all_aboard', data=comparison_total_fatalities, color="#6d83c9", alpha=0.2)
plt.plot(comparison_total_fatalities.index, 'all_aboard', data=comparison_total_fatalities, marker=".", color="#6d83c9", alpha=0.6, linewidth=1)
plt.fill_between(comparison_total_fatalities.index, 'total_fatalities', data=comparison_total_fatalities, color="#ff3333", alpha=0.2)
plt.plot(comparison_total_fatalities.index, 'total_fatalities', data=comparison_total_fatalities, color="#ff3333", marker=".", alpha=0.6, linewidth=1)
plt.legend(fontsize=10)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Amount of people', fontsize=10)
plt.title('Comparison: total fatalities vs all aboard per year', loc='Center', fontsize=14)
plt.tight_layout()
plt.grid(True)
plt.show()
```
